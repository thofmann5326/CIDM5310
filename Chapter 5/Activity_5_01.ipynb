## For this activity, all you will need is the pandas library. Load it into the first cell of the notebook.

```python
import pandas as pd
```

## Read in the mushroom.csv data from the Datasets directory and list the first five rows using .head().

```python
file_url = "https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter05/Datasets/mushroom.csv"

mushroom = pd.read_csv(file_url)

mushroom.head()
```

## You will see the class column and many visible attributes. List all the columns to see what else there is to work with. The result should be as follows:

```python
print(mushroom.columns)
```

## In addition to class, you see population and habitat, which are not visible attributes. You decide to create a multi-index using class, population, and habitat.

```python
my_index = pd.MultiIndex.from_frame(mushroom[['class', 'population', 'habitat']])
```

## Now, drop the columns that are in the index and set the DataFrame index to the multi-index. Make sure that you drop the existing default index.

```python
mushroom.drop(columns = ['class', 'population', 'habitat'], inplace = True)

mushroom.set_index(my_index, inplace = True)
mushroom.head(10)
```
