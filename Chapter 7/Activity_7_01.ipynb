## Import the pandas package.

```python
import pandas as pd
```

## Load the CSV file (using the ; delimiter to separate the columns) as a DataFrame.

```python
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter07/Data/student-mat.csv'

data_frame = pd.read_csv(file_url, delimiter=';')

data_frame
```

## Modify the DataFrame to contain only these columns: school, sex, age, address, heath, absences, G1, G2, and G3.

```python
data_frame = data_frame[['school', 'sex', 'age', 'address', 'health', 'absences', 'G1', 'G2', 'G3']]
```

## Display the first 10 rows of the DataFrame.

```python
data_frame.head(10)
```

## Build a pivot table that is indexed on school.

```python
pd.pivot_table(data_frame,index=['school'])
```

## Build a pivot table that is indexed on school and age.

```python
pd.pivot_table(data_frame,index=['school', 'age'])
```

## Build a pivot table that is indexed on school, sex, and age, with the mean and sum aggregation on the absences column.

```python
pd.pivot_table(data_frame,index=['school', 'sex','age'], values = 'absences', aggfunc={'mean', 'sum'})
```
